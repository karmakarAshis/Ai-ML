```python
# possitional args
def matt(a,b,c):
    print(f"{a}, {b}, {c}")



matt(10, 20, 30)
```

```python
def adder(*args):
    print(args)

adder(10, 20, 30, 30, 40, 50, 50)
```

```python
# keyword args
def matt(a,b,c):
    print(f"{a}, {b}, {c}")



matt(a=10, c=20, b=30)
```

```python
def adder(**kwargs):
    print(kwargs)

    for k,v in kwargs.items():
        print(f"{k} : {v}")


adder(emiiter=30, es=30, epoch=40, prob=50, omt=50)
```

generator

```python
lst = [x for x in range(500)]

def data_loader(chunk_size, lst):
    for i in range(0, len(lst), chunk_size):
        yield lst[i: i+chunk_size]

x = data_loader(5, lst)


print(next(x))
print(next(x))
```
