```python
# base/parent class
class Phone:
    category = "Electronics"


    # constractor
    def __init__(self, model, battery, camera, battery_percentage=100):
        self.model = model
        self.battery = battery
        self.camera = camera
        self.battery_percentage = battery_percentage

    # methods
    def charge(self, hour):
        print(f"charge complete by {hour}")

    def capture(self, photo):
        if(self.battery_percentage) <=0:
            print("no charge")
        else:
            self.battery_percentage -= photo
            print(f"photo capture in {self.model}")

```

1. Inheritance


```python
class Cooling_Mechanism:
    def __init__(self, cooling_methods):
        self.cooling_methods = cooling_methods
    
    def cooling_on(self):
        print(f"the system is being cooled by {self.cooling_methods}")
        
```

```python
# creating object
apple = Phone("Iphone 17", 3540, 48)

samsung = Phone("S25 Ultra", 4950, 108)

google = Phone("pixel 10", 4700, 50)
```

```python
apple.capture(10)
```

```python
apple.battery_percentage
```

```python
# child class
class SmartPhone(Phone):
    def __init__(self, model, battery, camera, processor):
        super().__init__(model, battery, camera)
        self.processor = processor
    
    def charge(self, hour):
        print("first charging in process")
        super().charge(hour)
```

```python
pro = SmartPhone("x", 5000, 100, "Snapdragon")
```

```python
pro.model
```

```python
pro.charge(10)
```

```python
pro.battery_percentage
```

```python
class SmartPhone_coolingMethod(SmartPhone, Cooling_Mechanism):
    def __init__(self, model, battery, camera, processor, cooling_method):
        SmartPhone.__init__(self, model, battery, camera, processor)
        Cooling_Mechanism.__init__(self,cooling_method)
```

```python
Pro_cooling = SmartPhone_coolingMethod("Y", 5000, 100, "SD", "Nitrogen")
print(Pro_cooling.processor)
print(Pro_cooling.battery)
print(Pro_cooling.cooling_methods)
print(Pro_cooling.cooling_on)
```

2. Polymorphism

```python
class Camera:
    def __init__(self, name):
        self.name =name
    
    def capture(self):
        print("A photo is captured")
```

```python
class Smart_phone(Camera):
    def __init__(self, name, resulation):
        super().__init__(name)
        self.resulation = resulation
    
    # method overiding
    def capture(self):
        print("Photo captured by Phone")
    

class DSLR(Camera):
    def __init__(self, name, resulation):
        super().__init__(name)
        self.resulation = resulation
    
    # method overiding
    def capture(self):
        print("Photo captured by DSLR")


class Drone(Camera):
    def __init__(self, name, resulation):
        super().__init__(name)
        self.resulation = resulation
    
    # method overiding
    def capture(self):
        print("Photo captured by Drone")
```

```python
phone = Smart_phone("Phone", 30)
dslr = DSLR("DSLR", 200)
drone = Drone("Drone", 150)
```

```python
phone.capture()
dslr.capture()
drone.capture()
```

3. Encapsulation

```python
class Mobile:
    def __init__(self, name, model, imei):
        self.__name = name
        self.__model = model
        self.__imei = imei # private

    def charge(self):
        print("charging")


    def imei_getter(self):
        return self.__imei
    
    def imei_setter(self, imei):
        self.__imei = imei
         

iphone = Mobile("Phone", "17", "1xlhdf")


print(iphone.imei_getter())


iphone.imei_setter("heqae")

print(iphone.imei_getter())
```

4. Abstraction

```python
from abc import ABC, abstractmethod
class Telephone(ABC):

    @abstractmethod
    def make_call(self):
        pass
    


class Sphone(Telephone):
    def make_call(self):
        print("Making a call using sphone")


class Iphone(Telephone):
    def make_call(self):
        print("Making a call using iphone")


```
